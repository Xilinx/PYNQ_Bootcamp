### Using If Statements

Sometimes, we want a statement to execute only if certain conditions are true. Just like in real life, I want to bring an umbrella outside only *if* it is raining. This is where an if statement comes in handy. 

If the condition we want is true, it executes the statements inside of it. If the condition is false, it will simply skip them, unless we include an *else* statement after, in which case it will execute the statements inside the else instead. 

You may have also noticed that we are using two equal signs in our comparison, instead of one. In general, one equal sign is used to assign values, like our variable `counter`, while a double equal sign is used for comparisons: Is x equal to y? 



````python
if nouns[0] == "small":
    print("True!")
else:
    print("False :(")
````

### Using For Loops

So far we've learned some basic Python statements, but it can get a little clunky if we're trying to execute multiple statements in a row. Next, we'll learn about how to condense our code using a loop.

We can use a `for` loop to iterate over all of the elements in a list. In simpler terms, our temporary variable `number` takes on each value in the list, executing the statements within the loop once as each value. 

````python
for number in [0, 1, 2, 3]:
    print(number)
    print("I'm repeating!")
````

Sometimes we would like to iterate through many numbers in our loop, and typing all of these out would be a big waste of time! Instead of supplying the list directly, let's use [range(...)](https://www.programiz.com/python-programming/methods/built-in/range) to make a list for us.

We tell `range(...)` how many numbers to give us and it will make a list by starting from 0 and counting up by 1 each time. For example, `range(4)` would give us 4 numbers starting from zero (`[0, 1, 2, 3]`).

Let's try it out.

````python
for number in range(4):
    print(number)
    print("I'm repeating!")
````

### Level up: hardware libraries (this is where the magic happens!)

The `random` library we just used is a **software library** — it is plain Python code that runs on the SOM board's regular processor (its CPU), just like the code on a laptop or a phone.

But your board is special. Underneath the Linux and Python you have been using, there is a big block of **reprogrammable hardware** called an **FPGA**. Some Python libraries don't just *run code* — they actually **rewire the chip** so it can do a job thousands of times faster than the CPU could on its own. We call these libraries (and the hardware designs they load) **overlays**.

The most exciting example of this is a hardware library that sets up a **DPU**. Let's break down what that means and what each line of setup code is really doing. 👇

#### First, what is an FPGA?

**FPGA** stands for **Field-Programmable Gate Array**. That is a mouthful, so here is the idea:

* A normal computer chip (a CPU) is like a **kitchen with one very fast chef**. You hand the chef a recipe (your program) and they do every step, one after another, super quickly.
* An **FPGA is like a box of LEGO bricks for building circuits**. Instead of one chef doing steps in order, you can *build a custom machine* that has hundreds of little workers all doing their piece **at the same time** (in parallel).

The word *field-programmable* means you can rebuild that machine **any time you want**, even after the chip is already in your board — you just load a new design. When we load an overlay in Python, we are literally handing the FPGA a new set of "LEGO instructions" and it rewires itself in a fraction of a second.

#### So what is a DPU?

**DPU** stands for **Deep-learning Processing Unit**. It is a special circuit design that we build *inside* the FPGA. Its whole job is to run **neural networks** (the "brains" behind modern AI) extremely fast.

A neural network — like an image classifier that can look at a photo and say "that's a cat" or "that's a school bus" — does a *huge* amount of multiply-and-add math. A CPU would grind through that math slowly. The DPU is a factory built specifically for that kind of math, with many workers running in parallel, so it can process images fast enough to feel "live."

The key idea:

> **The DPU is not a chip we buy separately — it is a design we *load into* the FPGA using a Python library.**

Now let's look at the actual setup calls, one at a time.

#### Step 1 — Import the DPU library

Just like we wrote `import random` to get the random-number tools, we import the tools that know how to talk to the DPU. On PYNQ boards this lives in a package called `pynq_dpu`, and the main tool is a class called `DpuOverlay`.

> **Note for the intro session:** the next few cells are *real* DPU commands. They only work once the AI board files are installed, so if a cell gives an error today, that's totally fine — the goal right now is to *understand* what each line does.

````python
from pynq_dpu import DpuOverlay
````

That single line loads a whole toolbox. `DpuOverlay` is a **blueprint** (programmers call it a *class*) that knows how to:

1. reprogram the FPGA with a DPU design, and
2. feed a neural network into that DPU.

Remember the `?` trick from the `random` library? It works here too. Run the cell below to read the built-in documentation for `DpuOverlay`.

````python
DpuOverlay?
````

#### Step 2 — Load the overlay (this is what reprograms the FPGA!)

Next we create a `DpuOverlay` object and hand it the name of a **bitstream file** (it usually ends in `.bit`). A bitstream is the file that contains the "LEGO instructions" for the FPGA — in this case, the instructions for building a DPU.

````python
overlay = DpuOverlay("dpu.bit")
````

**What just happened to the hardware?** A lot! When that line runs, PYNQ:

1. **Finds the bitstream** (`dpu.bit`) that describes the DPU circuit.
2. **Downloads it into the FPGA.** The chip physically *reconfigures* — its internal wiring changes so that a DPU now exists where there was blank hardware before. This takes only a fraction of a second.
3. **Sets up the connection** so your Python code and the new DPU can pass data back and forth (through shared memory).

In short: before this line, the FPGA was blank hardware; after it, there is a working AI accelerator inside your board. The variable `overlay` is now our remote control for that new hardware.

#### Step 3 — Load the AI model into the DPU

The DPU is the *factory*, but it still needs to know **what to build** — that is, which neural network to run. We load a trained model file (it ends in `.xmodel`). This file holds all the numbers the AI learned during training, packaged so the DPU can use them.

Here we'll use **ResNet-50**, a famous image-classification model that ships as a ready-to-run example with the `pynq_dpu` package. Any board with `pynq_dpu` installed already has `dpu_resnet50.xmodel`, so it's a great one to experiment with right away.

````python
overlay.load_model("dpu_resnet50.xmodel")
````

Notice the shape of this call: `overlay.load_model(...)`. Just like `random.randint(...)` was "the `randint` tool that lives inside `random`," here `load_model` is a tool that lives inside our `overlay`. The dot (`.`) means *"reach inside this thing and use one of its tools."*

After this line, the DPU is fully set up: the hardware is built **and** it knows which AI model to run.

#### Step 4 — Get the "runner" that does the work

Finally, we ask the overlay for a **runner** — the object we actually use to send data in and get answers out. Once we have it, we can push an image in and read the AI's prediction back.

````python
dpu = overlay.runner
````

#### Step 5 — Ask the DPU about its shape

Before feeding it pictures, it helps to ask the DPU what size of data it expects (its **input tensors**) and what size of answer it gives back (its **output tensors**). A *tensor* is just a fancy word for a grid of numbers — for an image, that's the width, height, and color channels.

````python
inputTensors  = dpu.get_input_tensors()
outputTensors = dpu.get_output_tensors()

print("The DPU expects images shaped like:", inputTensors[0].dims)
print("The DPU gives answers shaped like: ", outputTensors[0].dims)
````

#### Putting it all together

Here is the whole DPU setup in one place. Read it top to bottom and see if you can say, in your own words, what each line does to the board:

```python
from pynq_dpu import DpuOverlay          # 1. get the DPU tools
overlay = DpuOverlay("dpu.bit")          # 2. rewire the FPGA into a DPU
overlay.load_model("dpu_resnet50.xmodel")  # 3. give the DPU its AI brain
dpu = overlay.runner                     # 4. grab the remote control
inputTensors  = dpu.get_input_tensors()  # 5a. what data goes in?
outputTensors = dpu.get_output_tensors() # 5b. what answer comes out?
```

The big takeaway: **software libraries change what your code *does*; hardware libraries change what your chip *is*.** Real AI projects combine both — a little Python glue on the CPU, and a DPU doing the heavy AI lifting inside the FPGA.

## Talking to the real world: peripherals

So far everything we've done has stayed *inside* the board — printing text, making lists, even setting up the DPU. But one of the coolest things about the KV260 is that it can reach out and control **real hardware** you plug into it. These add-on devices are called **peripherals**.

Peripherals are things like buttons, sensors, buzzers, motors, and lights. Over the next few days you'll get to play with several of them. To warm up, let's control a **Grove LED bar** — a little stick of 10 LEDs — that is plugged into the board right now through the **PYNQ Grove Adapter**. 💡

### Step 1 — Connect to the LED bar

The `pynq_peripherals` library knows how to talk to lots of Grove devices. The LED bar plugs into the Grove adapter, which itself plugs into the board's **PMODA** connector. We tell the library that a **`grove_ledbar`** is on port **G2** of that adapter, and it hands us back an object we can control.

> If your LED bar is plugged into a different port, change `G2` to match (the ports are labelled on the adapter).

````python
from pynq_peripherals import PmodGroveAdapter

# `overlay` is the overlay we loaded earlier — it also gives us the PMOD port.
adapter = PmodGroveAdapter(overlay.PMODA, G2='grove_ledbar')
ledbar  = adapter.G2
````

### Step 2 — Turn some lights on!

The LED bar has 10 LEDs. We control it with:

`ledbar.set_level(level, brightness, green_to_red)`

* **level** — how many LEDs to light up (0 to 10)
* **brightness** — how bright they glow
* **green_to_red** — the direction the colors run

Let's light up **5** of them.

````python
brightness   = 3
green_to_red = 1

ledbar.set_level(5, brightness, green_to_red)   # light up the first 5 LEDs
````

Nice — go look at the board, you should see 5 lights glowing!

There's also a handy `clear()` command that turns the whole bar off.

````python
ledbar.clear()   # turn all the LEDs off
````

### Step 3 — Make it move with a loop

Remember `for` loops from earlier? We can combine a loop with the LED bar to make an animation. This lights the bar up one LED at a time, like a loading bar, then clears it.

````python
import time

# Fill up one LED at a time
for level in range(1, 11):     # 1, 2, 3, ... 10
    ledbar.set_level(level, brightness, green_to_red)
    time.sleep(0.1)

time.sleep(0.5)
ledbar.clear()                 # all off
````

You just made physical hardware respond to your code! This is the same idea behind everything from traffic lights to game controllers: software on the board sending signals out to real-world devices.
