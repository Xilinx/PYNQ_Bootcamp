# Remote board Setup

This notebook is used to setup a new board for remote access and trainings using remote boards.
The notebook adds a new systemd service to the board that reclones PYNQ Bootcamp git repo on each power cycle. 

This makes sure that each user session starts with a fresh clone.

Copy bash script to reclone training material and systemd service to the correct location.

```python
%%bash

sudo cp reclone_training.sh /usr/local/bin/reclone_training.sh
sudo chmod +x /usr/local/bin/reclone_training.sh

sudo cp reclone_training.service /etc/systemd/system/reclone_training.service
```

Setup systemd service

```python
%%bash

sudo systemctl daemon-reload
sudo systemctl enable reclone_training.service
```

Start systemd service

```python
%%bash

sudo systemctl start reclone_repo.service
```

Check systemd service status

```python
%%bash

sudo systemctl status reclone_repo.service
```
