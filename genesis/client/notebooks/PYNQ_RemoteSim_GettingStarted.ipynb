# Remote Robot Simulation with Genesis

This notebook shows you how to control a robot arm running on a remote server!

## What you'll learn:
1. Connect to the simulation server
2. Create a robot environment
3. Move the robot arm
4. Pick up objects
5. Record video of your robot

```python
# Import the simulation client
from pynqsim import SimulationClient

# Connect to the Genesis server (ask your instructor for the IP and port!)
SERVER_IP = "172.20.166.171"  # <-- Change this to your server's IP
SERVER_PORT = 9003           # <-- Change this to your server's port

sim = SimulationClient(SERVER_IP, SERVER_PORT)

print("Connected to server!")
```

## Create Your Robot Environment

Let's create an environment with a robot arm and some objects to play with.

```python
# Create environment with the pick_and_place scene
sim.create_environment(scene="pick_and_place")
print("Environment created!")
```

## Reset the Robot Arm

The robot arm can move to any position you want! 
- X axis: forward/backward
- Y axis: left/right  
- Z axis: up/down

```python
# Move the robot arm to its reset state
# Use smooth motion for safe, collision-free movement
sim.move_robot(0, position=[0.0, 0.0, 0.0], smooth=True, num_waypoints=100)

print("Robot moved!")
```

## Move the Robot Arm

The robot arm can move to any position you want! 
- X axis: forward/backward
- Y axis: left/right  
- Z axis: up/down

```python
# Move the robot arm to hover above the table
# Use smooth motion for safe, collision-free movement
sim.move_robot(0, position=[0.5, 0.0, 0.43], smooth=True, num_waypoints=100)

print("Robot moved!")
```

```python
# Check where the robot is
state = sim.get_state(0)
print(f"End effector position: {state['end_effector']['position']}")
```

## Pick Up an Object

Let's pick up one of the cubes on the table!

```python
# Move above the red cube
sim.move_robot(0, position=[0.4, -0.1, 0.55], smooth=True, num_waypoints=60)

# Open gripper
sim.open_gripper(0)
sim.step(50)

# Lower slowly to just above the cube (table top ~0.41, cube top ~0.45)
sim.move_robot(0, position=[0.4, -0.1, 0.47], smooth=True, num_waypoints=50)

# Close gripper to grab
sim.close_gripper(0)
sim.step(100)

# Lift the cube
sim.move_robot(0, position=[0.4, -0.1, 0.6], smooth=True, num_waypoints=40)

print("Cube picked up!")
```

## Record a Video

Let's record a video of our robot in action!

```python
# Start recording
sim.start_recording()

# Do some cool moves
sim.move_robot(0, position=[0.5, 0.1, 0.5])
sim.step(100)
sim.move_robot(0, position=[0.6, -0.1, 0.5])
sim.step(100)
sim.move_robot(0, position=[0.5, 0.0, 0.6])
sim.step(100)

# Stop recording and get video
video = sim.stop_recording()
print(f"Recorded {len(video)} bytes of video!")
```

```python
# Display the video
sim.show_video(video)
```

## Clean Up

Always destroy your environment when you're done!

```python
sim.destroy()
print("Environment destroyed. Goodbye!")
```
